```python
golden_data_set = []
for name in ['100', '200', '300', '400', '500', '600', '700']:
    data = np.genfromtxt('../data/golden/'+name+'.log', delimiter=',')
    golden_data_set.append(data[:,0:100])
```

```python
error_data_set = []
for name in ['100', '200', '300', '400', '500', '600', '700']:
    data = np.genfromtxt('../data/rate_16/'+name+'.log', delimiter=',')
    error_data_set.append(data[:,0:100])
error = np.array(golden_data_set) - np.array(error_data_set)
```

```python
check_error_set = []
for name in ['100', '200', '300', '400', '500', '600', '700']:
    data = np.genfromtxt('../data/checkpoint/'+name+'.log', delimiter=',')
    check_error_set.append(data[:,0:100])
check_point_error = np.array(golden_data_set) - np.array(check_error_set)
```

```python
plt.imshow(golden_data_set[-1])
plt.colorbar()
plt.show()
```

```python
plt.imshow(error[-1],cmap='Reds')
plt.colorbar()
plt.show()
```

```python
plt.imshow(check_point_error[-1],cmap='Reds')
plt.colorbar()
plt.show()
```

```python
plt.imshow(error[-1] - check_point_error[-1], cmap='Reds')
plt.colorbar()
plt.clim(0,0.00002)
```
